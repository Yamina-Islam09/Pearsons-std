```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1=pd.read_csv("transport.csv")
df1.head(13)
```

```python
from sklearn.preprocessing import OrdinalEncoder
#for tMode
ord_enc = OrdinalEncoder()
df1["tMode-1"] = ord_enc.fit_transform(df1[["tMode"]])
df1[["tMode", "tMode-1"]].head(11)

#for transport

ord_enc = OrdinalEncoder()
df1["transport-1"] = ord_enc.fit_transform(df1[["transport"]])
df1[["transport", "transport-1"]].head(11)

# for male,travelCost,income
ord_enc = OrdinalEncoder()
df1["gender-1"] = ord_enc.fit_transform(df1[["gender"]])
df1[["gender", "gender-1"]].head(11)

ord_enc = OrdinalEncoder()
df1["income-1"] = ord_enc.fit_transform(df1[["income"]])
df1[["income", "income-1"]].head(11)

ord_enc = OrdinalEncoder()
df1["travelCost-1"] = ord_enc.fit_transform(df1[["travelCost"]])
df1[["travelCost", "travelCost-1"]].head(11)

df1
```

```python
df2=df1.drop(['gender','travelCost','income','transport','tMode'],axis=1)
df2
# df1.drop(df1.column[0,1,2],axis=1)
```

```python
corr = df2.corr(method="pearson")

print(corr)
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            # print(i,j)
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                # print(colname)
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(df2, 0.18)
len(set(corr_features))
```
